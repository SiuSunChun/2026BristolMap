```python
import geopandas as gpd
from pyproj import Proj
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import EngFormatter
from matplotlib import rc
formatter0 = EngFormatter(unit=chr(176))
```

```python
url = "https://raw.githubusercontent.com/SiuSunChun/2026BristolMap/refs/heads/main/Bristol_boundary.geojson"
```

```python
# Load files
Map=gpd.read_file(url)

# Map.crs
bristol= Map.to_crs("EPSG:4326")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=1, sharex='all',figsize=(3,3),dpi=200)

axes = bristol['geometry'].plot(ax=axes, figsize=(2,3),linewidth=0.5, edgecolor='grey',alpha=1, facecolor='blue')

axes.tick_params(direction='in', labelsize=8)
axes.set_xlabel('Longitude', fontsize=7)
axes.set_ylabel('Latitude', fontsize=7)
axes.xaxis.set_major_formatter(formatter0)
axes.yaxis.set_major_formatter(formatter0)
fig.tight_layout() 
plt.show()
```
