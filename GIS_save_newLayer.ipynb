```python
import geopandas as gpd
from pyproj import Proj
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import EngFormatter
from matplotlib import rc
formatter0 = EngFormatter(unit=chr(176))
```

```python
# shape file location
path='https://raw.githubusercontent.com/SiuSunChun/2026BristolMap/refs/heads/main' # change the path 
```

```python
# read shape file
Map=gpd.read_file(path+'/UK.geojson')
# Map.crs # check the CRS
uk_bd= Map.to_crs("EPSG:4326") # Covert to desire CRS
```

```python
Map.to_file('UK.geojason')
```

```python
# select feature
c_ne=['Northern Ireland','Scotland',' Wales']
England=uk_bd[~uk_bd.iloc[:,1].isin(c_ne)]
```

```python
# Plot original file and new layer
fig,axes=plt.subplots(nrows=1, ncols=2,sharey=True,sharex=True, figsize=(6,4),dpi=300)

ax1 = England['geometry'].plot(ax=axes[1] , linewidth=0.5, edgecolor='grey',alpha=1, facecolor='blue')
ax2 = uk_bd['geometry'].plot(ax=axes[0], linewidth=0.5, edgecolor='grey',alpha=1, facecolor='none')

ax1.set_xticks(np.arange(-9, 2, step=5))
ax1.set_yticks(np.arange(49, 61, step=5))


ax1.set_title('England',fontsize=9)
ax1.tick_params(direction='in', labelsize=8)
ax1.set_ylabel('Latitude', fontsize=7)

ax2.set_title('UK Map',fontsize=9)
ax2.tick_params(direction='in', labelsize=8)
ax2.set_xlabel('Longitude', fontsize=7)
ax2.set_ylabel('Latitude', fontsize=7)
ax2.xaxis.set_major_formatter(formatter0)
ax2.yaxis.set_major_formatter(formatter0)
fig.tight_layout() 
plt.show()
```

```python
#save the new layer as SHP with desire CRS
England.to_crs(Map.crs)
England.to_file('England.geojason')
```
